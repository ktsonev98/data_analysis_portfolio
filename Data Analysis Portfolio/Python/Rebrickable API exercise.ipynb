```python
def fetch_sets_by_theme(theme_id, theme_name):
    all_results = []
    url = "https://rebrickable.com/api/v3/lego/sets/"
    params = {"theme_id": theme_id, "page_size": 100}

    while url:
        r = requests.get(url, headers=headers, params=params)
        data = r.json()
        for item in data["results"]:
            item["theme"] = theme_name  # label rows
        all_results.extend(data["results"])
        url = data["next"]
        params = None

    return pd.DataFrame(all_results)


df_star_wars = fetch_sets_by_theme(star_wars_id, "Star Wars")
df_bionicle  = fetch_sets_by_theme(bionicle_id, "Bionicle")

df_sets = pd.concat([df_star_wars, df_bionicle], ignore_index=True)
```

```python


def get_descendant_theme_ids(df_themes, root_id):
    children_map = df_themes.groupby("parent_id")["id"].apply(list).to_dict()

    stack = [root_id]
    all_ids = {root_id}

    while stack:
        current = stack.pop()
        for child in children_map.get(current, []):
            if child not in all_ids:
                all_ids.add(child)
                stack.append(child)

    return sorted(all_ids)


star_wars_theme_ids = get_descendant_theme_ids(df_themes, star_wars_id)
print("Star Wars theme IDs:", len(star_wars_theme_ids))


df_star_wars = fetch_sets_for_theme_ids(star_wars_theme_ids, "Star Wars", headers)
```

Theme-like columns exist that cause the discrepancy

```python


# 1) Find which "theme-like" column exists
theme_col = None
for c in ["theme", "theme_group", "Theme", "theme_name"]:
    if c in df_sets.columns:
        theme_col = c
        break

# Fallback: grab any column that contains the word "theme"
if theme_col is None:
    candidates = [c for c in df_sets.columns if "theme" in c.lower()]
    if candidates:
        theme_col = candidates[0]

if theme_col is None:
    raise ValueError(f"No theme column found. Available columns: {list(df_sets.columns)}")

# 2) Standardise it to "theme"
if theme_col != "theme":
    df_sets = df_sets.rename(columns={theme_col: "theme"})

# 3) Clean num_parts (safe even if already clean)
df_sets["num_parts"] = pd.to_numeric(df_sets["num_parts"], errors="coerce")
df_sets = df_sets.dropna(subset=["num_parts"])
df_sets = df_sets[df_sets["num_parts"] > 0]

# 4) Dedupe by set_num (important with subthemes)
if "set_num" in df_sets.columns:
    df_sets = df_sets.drop_duplicates(subset=["set_num"])

# 5) Checks
print("Theme column used:", theme_col, "-> standardised to 'theme'")
print("Max parts:", df_sets["num_parts"].max())
print(df_sets.groupby("theme")["set_num"].nunique())

df_sets.nlargest(10, "num_parts")[["theme", "set_num", "name", "year", "num_parts"]]

```
