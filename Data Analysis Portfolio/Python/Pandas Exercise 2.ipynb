```python
import matplotlib.pyplot as plt

import pandas as pd

df = pd.read_csv(r"student - for Pandas Exercise 1 and 2.csv")

df.info() 

df.head()
```

```python

top10 = df.sort_values(by="mark", ascending=False).head(10)

top10.plot.barh(x="name", y="mark", figsize=(10,6), title="Top 10 Students by Mark")

```

```python
avg_mark = df.groupby("class")["mark"].mean()

avg_mark.plot.bar(figsize=(10,6), title="Average Mark per Class")

plt.xlabel("Class")

plt.ylabel("Average Mark")

plt.show()
```
