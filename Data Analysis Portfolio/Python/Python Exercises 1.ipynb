Let's play a “guess the number game” using while loop.

Write a program that:
1. Takes the number from user
1. Compares the number the user entered with a randomly generated number
1. Provide a message “Little lower” or “Little higher” till the user guesses it correctly

```python
import random

user_input = int(input("I have picked a number between 10 and 99, can you guess what it is? "))
number_to_guess = random.randint(10, 99)

while user_input != number_to_guess:
    if user_input < number_to_guess:
        user_input = int(input("Little higher! "))
    else:
        user_input = int(input("Little lower! "))

print("Bingo!")

```
